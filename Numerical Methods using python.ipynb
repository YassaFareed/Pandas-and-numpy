# Bisection Method

```python
#solving xsinx-1
import math
def func(x):
     return x*math.sin(x)-1 # enter your function here 
def bisection1(a,b,sig):
    acc=1/sig
    if((func(a)>0 and func(b)>0) or (func(a)<0 and func(b)<0)):
        print("Wrong a and b are choosen")
        return
    c=a
    counter=1
    while((b-a)>=acc):
        c=(a+b)/2
        print("Iteration " ,counter,"= %.6f"%c)
        counter=counter+1
        if(func(c)>0):
            if(func(a)>0):
                a=c
            elif(func(b)>0):
                b=c
        elif(func(c)<0):
            if(func(a)<0):
                a=c
            elif(func(b)<0):
                b=c 
                
    print("Value of root: ","%.6f"%c)
    
accuracy=int(input("What is the solution accuracy you want to calculate (1 to 6) ? \n"))
sig=1
while(accuracy):
    sig=sig*10
    accuracy=accuracy-1
a=float(input("Enter first number :"))
b=float(input("Enter second number :"))
bisection1(a,b,sig)

    
   
```

# Fixed-point Method

```python
#solving (cosx+3)/2
import math
def func(x):
     return (math.cos(x)+3)/2    # enter your function here
def truncate(n, decimals=0):
    multiplier = 10 ** decimals
    return int(n * multiplier) / multiplier
def fixedpoint(a,accuracy):
    prev=0
    while(truncate(prev,accuracy)!=truncate(a,accuracy)):
        print("Iteration %.6f"%a) 
        prev=a
        a=func(a)
        
    print("The value of root is %.6f"%a)  
accuracy=int(input("Accuracy to which you want to calculate ? "))

a=float(input("Enter any point in range :"))
fixedpoint(a,accuracy)
```

```python
def newtonR(a,accuracy):
    prev=1.0 #default value
    variable=0.0 #default value
    while(truncate(variable,accuracy)!=truncate(prev,accuracy) or prev==1):
        if(prev==1):
            prev=a
        print("\n\n Iteration %.6f"%prev) 
        variable=prev
        val=func(prev)
        der=derivative(func,prev)
        der=der-2
        next_=variable-(val/der)
        prev=next_
    print("The value is %.6f"%prev)


    
accuracy=int(input("Accuracy to which you want to calculate ? "))
a=float(input("enter the initial value of x:"))
newtonR(a,accuracy)

    
```

# RK Method of order 4

```python
#just change the function and h value and change parameter of Y to get answer
h=0.1
import math
def func(t,y):
    y1=(y/t)-((y/t)**2) # add function here
    return y1
#def func(t,y):
#    y1=(t*math.exp(3*t))-(2*y)
#    return y1
def k1(t,y):
    K1=h*(func(t,y))
    return K1
def k2(t,y):
    y=y+(k1(t,y)/2)
    t=t+h/2
    K2=h*(func(t,y))
    return K2
def k3(t,y):
    y=y+(k2(t,y)/2)
    t=t+h/2
    K3=h*(func(t,y))
    return K3
def k4(t,y):
    y=y+(k3(t,y))
    t=t+h
    K4=h*(func(t,y))
    return K4
def RKMethod(t,y):
    a=k1(t,y)
    print("K1=",a)
    b=k2(t,y)
    print("K2=",b)
    c=k3(t,y)
    print("K3=",c)
    d=k4(t,y)
    print("K4=",d)
    ans=(a+2*b+2*c+d)/6
    ans=y+ans
    return ans

def hvalue(H):
    global h
    h=H
    
def Actual(t): #actual function
    x=t/(1+(math.log(t))) #change actual function here
    return x

#enter the function in actual function and HenzMethod function yourself
t=int(input("Enter initial value of t :"))
y=int(input("Enter initial value of y :"))
H=float(input("Enter h :"))
rangeofT=float(input("Enter starting range of t :"))
rangeofTend=float(input("Enter ending range of t :"))
hvalue(H)
iterations=(rangeofTend-rangeofT)/H
for i in range(int(iterations)):
    print("---{} iteration--- ".format(i+1))
    y=RKMethod(t,y)
    print("Estimated :",y)
    t=t+H
    print("Actual :",Actual(t))
    print()
    
```
